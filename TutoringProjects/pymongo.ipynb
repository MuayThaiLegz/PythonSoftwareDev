```python
import pymongo
from pymongo import MongoClient
from pymongo.errors import DuplicateKeyError
import hashlib

def make_hashes_len(password):
    
    return hashlib.sha256(password.encode()).hexdigest()[:8]
from random import randint

# Create a new user
newUserName = "Alice"
newUserEmail = "alice@example.com"
newUserNumber = "1234567890"
newUserType = "Passenger"
lat_point = randint(-90, 90)
lon_point = randint(-180, 180)
newUserId = make_hashes_len(newUserEmail)

def create_user_table():
    UserData.create_index([('newUserEmail', pymongo.ASCENDING)], unique=True)
    UserData.create_index([('newUserNumber', pymongo.ASCENDING)], unique=True)
    UserData.create_index([('newUserId', pymongo.ASCENDING)], unique=True)

def add_user_data(newUserName,newUserEmail, newUserNumber,newUserType ,lat_point,lon_point ,newUserId):
    try:
        UserData.insert_one({
            'newUserName': newUserName,
            'newUserEmail': newUserEmail,
            'newUserNumber': newUserNumber,
            'newUserType': newUserType,
            'lat_point': lat_point,
            'lon_point': lon_point,
            'newUserId': newUserId
        })
    except DuplicateKeyError as e:
        print(str(e))
        
def view_all_user():
    return list(UserData.find())

def edit_user_data(newUserId,newUserEmail,newUserNumber,newUserType):
    UserData.update_one({'newUserId': newUserId}, {'$set': {
        'newUserEmail': newUserEmail,
        'newUserNumber': newUserNumber,
        'newUserType': newUserType,
    }})
    
def delete_user_data(newUserId):
    UserData.delete_one({'newUserId': newUserId})
```
