```python
class LoanBuilder:
    
    def __init__(self, applicant):
        
        self.loan_amount = 25000 
        self.loan_duration_months = 33 
        self.LoanType = LoanEnum
        
        self.rate = 3.3/100/12
        
        self.payment = 1 * self.loan_duration_months
        
        self.paymentAmount = self.loan_amount / 1-(1 + self.rate)**-self.payment
        
        applicant.BorrowerDetails.update({"Payment_Amount" : self.paymentAmount})
        
        
        if self.loan_duration_months <= 18:
                applicant.BorrowerDetails.update({"LoanType": LoanType.ShortTerm})
                
        if self.loan_duration_months <= 30:
                applicant.BorrowerDetails.update({"LoanType" : LoanType.MidTerm})
        
        elif self.loan_duration_months >= 30:
                applicant.BorrowerDetails.update({"LoanType" : LoanType.LongTerm})

     
    def __str__(self, applicant):
        return applicant.BorrowerDetails
```

```python


# Database Funtions
def create_usertable():
	c.execute('CREATE TABLE IF NOT EXISTS userstable(username TEXT, password TEXT, address TEXT, location TEXT, email TEXT, img TEXT)')

def create_usergallery():
	c.execute('CREATE TABLE IF NOT EXISTS usergallery(username TEXT, DeeDOwner TEXT, DeeDEtails TEXT, DeeD_hash TEXT, topic TEXT, forSaleSpec TEXT, DeeD_array TEXT)')

def create_Transfers():
	c.execute('CREATE TABLE IF NOT EXISTS Transfers(newParty TEXT, hostParty TEXT, TransferDetails TEXT, TransferHash TEXT)')

def add_userdata(username, password, address, location, email, img):
	c.execute('INSERT INTO userstable(username,password,address,location,email,img) VALUES (?,?,?,?,?,?)',
	          (username, password, address, location, email, img))
	conn.commit()


def add_userlisting(username, DeeDOwner, DeeDEtails, DeeD_hash, topic, forSaleSpec, DeeD_array):
	c.execute('INSERT INTO usergallery(username,DeeDOwner, DeeDEtails, DeeD_hash, topic, forSaleSpec, DeeD_array) VALUES (?,?,?,?,?,?,?)',
				(username, DeeDOwner, DeeDEtails, DeeD_hash, topic, forSaleSpec, DeeD_array))
	conn.commit()


def add_Transferdata(newParty, hostParty, TransferDetails, tokenId, transferHash):
	c.execute('INSERT INTO Transfers(newParty, hostParty, TransferDetails, tokenId, TransferHash) VALUES (?,?,?,?,?)',
	          (newParty, hostParty, TransferDetails, tokenId, transferHash))
	conn.commit()





def login_user(username, email, password):
	c.execute('SELECT * FROM userstable WHERE username =? AND email = ? AND password=?', (username, email, password))
	Logindata = c.fetchall()
	
	return Logindata


def view_all_users():
	c.execute('SELECT username, address, email, location, img FROM userstable')
	data = c.fetchall()

	return data


def view_all_listings():
	c.execute('SELECT username, DeeDOwner, DeeDEtails, DeeD_hash, topic, forSaleSpec, DeeD_array FROM usergallery')
	Listingdata = c.fetchall()

	return Listingdata

def view_all_Transfers():
	c.execute('SELECT newParty, hostParty, TransferDetails, TransferHash FROM Transfers')
	Transfersgdata = c.fetchall()

	return Transfersgdata


def view_user_listing(username, password):
	c.execute('SELECT * FROM userstable AND usergallery WHERE username =? AND password = ?', (username, password))
	Listing_data = c.fetchall()

	return Listing_data


def convertToBinaryData(filename):
    with open(filename, 'rb') as file:
        blobData = file.read()
    return blobData



def make_hashes(password):
	return hashlib.sha256(str.encode(password)).hexdigest()


def hash_users(username):
	return hashlib.sha256(str.encode(username)).hexdigest()



def check_hashes(password, hashed_text):
	if make_hashes(password) == hashed_text:
		return hashed_text
	return False


def check_users(username, hash_users):
	if hash_users(username) == hash_users:
		return hash_users
	return False

```
